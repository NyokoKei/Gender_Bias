```python
import pandas as pd
```

```python
#https://github.com/mledoze/countries/tree/master
df = pd.read_json('source_data/countries.json')
df
```

```python
df = df[df['unMember'] == True]
df
```

```python
df = df.loc[:, ['name', 'demonyms']]
df
```

```python
df.iloc[0, 0]
```

```python
df.iloc[4, 1]
```

```python
unique_keys = set(key for d in df['demonyms'] for key in d.keys())
print(unique_keys)

filtered_rows = df[df['demonyms'].apply(lambda x: x['eng']['f'] != x['eng']['m'])]

filtered_rows
```

```python
df['name'] = df['name'].apply(lambda x: x['common'])
df
```

```python
df['demonym'] = df['demonyms'].apply(lambda x: x['eng']['f']) #since for eng m and f are similar
df
```

```python
df.reset_index(inplace=False)
df
```

```python
df2 = df.loc[:, ['name', 'demonym']]
df2 = df2[df2.name != 'Vatican City']
df2.to_csv('output_data/v2_UN_countries.csv', index=False)
```

```python
import pandas as pd

# Path to your text file containing the HTML table
html_file = 'untitled.txt'

try:
    # Read the HTML content from the file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    # Use pandas to read HTML tables
    tables = pd.read_html(html_content)
    
    # Check how many tables were found
    print(f"Number of tables found: {len(tables)}")
    
    # Assuming the desired table is the first one
    df = tables[0]
except FileNotFoundError:
    print(f"The file '{html_file}' does not exist.")
except ValueError as ve:
    print(f"Error parsing tables: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
