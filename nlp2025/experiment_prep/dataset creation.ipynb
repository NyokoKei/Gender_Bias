# French

```python
fra = pd.read_csv('../output_data/France.csv')
fra['fra_m'] = fra['fra_m'].apply(lambda x: x.split(' et ')[0])
fra['fra_f'] = fra['fra_f'].apply(lambda x: x.split(' et ')[0])
fra.head()
```

```python
fra['fra_sent'] = fra.apply(lambda row: {'m': 'Je suis ' + row['fra_m'] + '.', 'f': 'Je suis ' + row['fra_f'] + '.', 'n': None}, axis=1)
fra
```

```python
data = fra[['name', 'fra_sent']]
data
```

```python
data = data.merge(df, on='name', how='inner')
data
```

```python
data.columns
```

```python
data['sentence'] = data.apply(lambda row: {'eng': row['sentence'], 'fra': row['fra_sent']}, axis=1)
data = data.drop('fra_sent', axis=1)
data
```

```python
data.to_csv('output/dataset.csv', index=False)
```
