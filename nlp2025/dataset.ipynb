```python
html_file = 'source_data/german.txt' #https://deutsch.lingolia.com/en/vocabulary/laender-nationalitaeten
try:
    # Read the HTML content from the file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    # Use pandas to read HTML tables
    tables = pd.read_html(html_content)
    
    # Check how many tables were found
    print(f"Number of tables found: {len(tables)}")
    
    # Assuming the desired table is the first one
    df = tables[0]
except FileNotFoundError:
    print(f"The file '{html_file}' does not exist.")
except ValueError as ve:
    print(f"Error parsing tables: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
df.head()
```

```python
df['Country in English'] = df['Country in English'].apply(lambda x: x.split('(')[0]) 
df = df.rename(columns={'Country in English': 'name'})
df = df.drop(['Country in German', 'Adjective'], axis=1)
df.sample(5)
```

# Spanish

```python
import pandas as pd
```

```python
# https://espanol.lingolia.com/en/vocabulary/countries
html_file = 'source_data/spanish.txt' #https://espanol.lingolia.com/en/vocabulary/countries
try:
    # Read the HTML content from the file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    # Use pandas to read HTML tables
    tables = pd.read_html(html_content)
    
    # Check how many tables were found
    print(f"Number of tables found: {len(tables)}")
    
    # Assuming the desired table is the first one
    df = tables[0]
except FileNotFoundError:
    print(f"The file '{html_file}' does not exist.")
except ValueError as ve:
    print(f"Error parsing tables: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
df
```

```python
df.columns
```

```python
df['spa_m'] = df['Demonym*/Adjective'].apply(lambda x: x.split('/')[0] if '/-' in x else None)
df['spa_n'] = df['Demonym*/Adjective'].apply(lambda x: x if '/' not in x or '-' not in x else None) 
df
```

```python
df.to_csv('temp.csv', index=False)
```

```python
def feminine(demonym):
    '''
    '''
    if '/-' not in demonym:
        return None
    else:
        demonym = demonym.split(' ')
        demonym = demonym[0]   
    m, f = demonym.split('/-')
    # print(m, f)
    # print(m[-1] == 's', f == 'sa')
    if (m[-1] == 's' and f == 'sa'):
        return m[:-1]+f
    elif(m[-2:] == 'io'):
        return m[:-3]+f
    elif(m[-1] == 'o'):
        return m[:-2]+f
```

```python
df['spa_f'] = df['Demonym*/Adjective'].apply(lambda x: feminine(x))
```

```python
df
```

```python
df[df.spa_n == df.spa_f]
```

```python
df = df.rename(columns={'Country in English': 'name'})
df = df[['name', 'spa_m', 'spa_f', 'spa_n']]
df
```

```python
df.to_csv('output_data/Spain.csv', index=False)
```

# Italian

```python
# https://www.theintrepidguide.com/nationalities-in-italian/?utm_source=chatgpt.com
html_file = 'source_data/italian.txt' 
try:
    # Read the HTML content from the file
    with open(html_file, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    # Use pandas to read HTML tables
    tables = pd.read_html(html_content)
    
    # Check how many tables were found
    print(f"Number of tables found: {len(tables)}")
    
    # Assuming the desired table is the first one
    df = tables[0]
except FileNotFoundError:
    print(f"The file '{html_file}' does not exist.")
except ValueError as ve:
    print(f"Error parsing tables: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
df
```

```python
it = pd.DataFrame(columns=['name', 'it_m', 'it_f', 'it_n'])
it['name'] = df['Country (English)']
it['it_m'] = df['Nationality (Masc. Sing.)']
it['it_f'] = df['Nationality (Fem. Sing.)']
it
```

```python
it['it_n'] = it.apply(lambda row: row['it_m'] if row['it_m'] == row['it_f'] else '', axis=1)
it['it_m'] = it.apply(lambda row: '' if row['it_n'] != '' else row['it_m'], axis=1)
it['it_f'] = it.apply(lambda row: '' if row['it_n'] != '' else row['it_f'], axis=1)
it
```

```python
it['it_sent'] = it.apply(lambda row: {'m': 'Sono ' + row['it_m'] + '.' if row['it_m'] != '' else None, 
                                      'f': 'Sono ' + row['it_f'] + '.' if row['it_f'] != '' else None, 
                                      'n': 'Sono ' + row['it_n'] + '.' if row['it_n'] != '' else None}, axis=1)
it
```

```python
it.to_csv('output_data/Italy.csv', index=False)
```
